Load and Inspect

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")

iris.head()
```

```python
print(iris.columns)
print(iris.shape)
```

```python
penguins.head()
```

```python
print(penguins.columns)
print(penguins.shape)
```

Class Distribution

```python
plt.figure()
sns.countplot(x="species", data=iris)
plt.title("Iris Species Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x="species", data=penguins)
plt.title("Penguins Species Count")
plt.show()
```

Feature Distribution

```python
plt.figure()
sns.histplot(iris["sepal_length"], bins=20, kde=True)
plt.title("Sepal Length Distribution")
plt.show()
```

```python
plt.figure()
sns.histplot(iris["petal_length"], bins=20, kde=True)
plt.title("Petal Length Distribution")
plt.show()
```

Scatter Plot

```python
plt.figure()
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

Pair Plot

```python
sns.pairplot(iris, hue="species")
plt.show()
```

Heatmap

```python
plt.figure()
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Iris")
plt.show()
```

Box Plot

```python
plt.figure()
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length by Species")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Petal Length by Species")
plt.show()
```

Violin Plot

```python
plt.figure()
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Violin Plot - Sepal Length")
plt.show()
```

```python
plt.figure()
sns.violinplot(x="species", y="petal_length", data=iris)
plt.title("Violin Plot - Petal Length")
plt.show()
```

Compare Datasets

```python
plt.figure()
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
plt.title("Iris: Sepal vs Petal Length")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins)
plt.title("Penguins: Flipper Length vs Body Mass")
plt.show()
```
